# Creating DataFrame

## Creating a dictionary

```python
import pandas as pd
```

data = {'Animals':['cow','chicken','rabbit','goats'],
        'Age':[4,1,2,3],
        'Cost':[20000,2000,4000,10000]}
df=pd.DataFrame (data)
df

```python
Data = [['cow',4,20000],
        ['chicken',1,2000],
        ['rabbit',2,4000],
        ['goats',3,10000]]
df=pd.DataFrame (data,columns= ['Animals','Age','Cost'])
df
```

## From a Numpy Array

```python
import numpy as np
```

```python
arr = np.array ([[1,2,3],[4,5,6]])
df=pd.DataFrame (arr,columns=['A','B','C'])
print(df)
                
```

## From a dictionary with lists (index)

```python
data = {'Name':['John','Victor','Edwin'],
        'Age':[20,19,17]}
df=pd.DataFrame (data, index=['Row1','row2','Row3'])
df
```
